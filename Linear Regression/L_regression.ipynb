Linear Regression

```python
def linear_regression(x,y):
    w0=random.uniform(0.0,1.0)
    w1=random.uniform(0.0,4.75)
    alpha=.001
    
    for i in range(200):
        slope_w0=(y-(w0+w1*x))*(-1)
        slope_w1=(y-(w0+w1*x))*(-x)
        
        w0=w0-alpha*slope_w0    
        w1=w1-alpha*slope_w1
        
    return w0,w1
        
```

```python
w0,w1 =linear_regression(x,y)
predict_y=w0+(w1*x)
color='blue'
plt.scatter(x,predict_y,c='red')
plt.scatter(x,y,c=color)
plt.xlabel("Population")
plt.ylabel("Profit")
```
