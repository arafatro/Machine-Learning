```python
def do_logistic_regression(x, y):
    w0 = 0.5 # just start with an arbitrary value
    w1 = 1
    w2 = 1
    alpha = 0.001
    n = len(x)
    
    cost = []
    
    for i in range(10000):
        slope_wrt_w0 = sum((y - (w0 + w1*x)) * -1) / n
        slope_wrt_w1 = sum((y - (w0 + w1*x)) * -x) / n
        
        w0 = w0 - alpha * slope_wrt_w0
        w1 = w1 - alpha * slope_wrt_w1
        
        cost.append(sum((y - (w0 + w1 * x))**2)/n)
    
    print('dJ/dw0:', slope_wrt_w0)
    print('dJ/dw1:', slope_wrt_w1)
    
    return w0, w1, cost
        
```

```python
w0, w1, cost = do_logistic_regression(x[:, 0], x[:, 1])
predicted_y = w0 + w1 * x[:, 0]
```
