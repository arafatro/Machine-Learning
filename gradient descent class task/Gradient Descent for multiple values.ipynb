```python
def linear_regression(x,y,no_of_x):
    w0=0.5
    w=[]
    x_column=[]
    
    for i in range(no_of_x):
        x_column.append(x[:,i])
        w.append(0.5)
    
    np_x=np.array(x_column)
    np_w=np.array(w)
    length=len(x)/2
    alpha=.001
    k=[]

    for i in range(1000):
        for j in range(no_of_x):
            #k.append(sum((y-(w0+(w[j]*np_x[j])**2))))
            predict_y =(w0+(np_w[j]*np_x[j]))/length
            slope_w0 =sum((y-predict_y)*(-1))
            slope_w =sum((y-predict_y)*(-np_x[j]))
            w0=w0-(alpha*slope_w0)
            np_w=np_w[j]-(alpha*slope_w)
      
    return predict_y,np_x
   
```
