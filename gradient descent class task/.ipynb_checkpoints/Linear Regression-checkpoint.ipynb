```python
import pandas as pn
data=pn.read_csv("MLinear.csv",header=None)
data
```

```python
import numpy as np
data=np.asarray(data)
data.shape
```

```python
X=data[:,:data.shape[1]-1]
y=data[:,-1:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,:1],y[:,:1])
plt.xlabel("Population")
plt.ylabel("Profit")
```

```python
w0=3.0
w1=1.8
alpha=0.001
predicted_Y=w0+w1*X
plt.scatter(X[:,:1],y[:,:1])
plt.xlabel("Population")
plt.ylabel("Profit")
plt.scatter(X[:,:1],predicted_Y[:,:1])
```

```python
w0=0.5
w1=0.5
alpha=0.0001

maxIter=200

for i in range(maxIter):
    # in each iteration calculate slope and update weights
    
    #step1: lets find the predicted y first
    predictedY=w0+w1*X
    #print(predictedY)
    
    # step 2: Lets find the error or difference between predictedY and realY
    
    error = y-predictedY
    
    # step 3 : find slope 0
    
    slope0 = sum(error*(-1))
    
     # step 4 : find slope 1
    slope1 = sum(error*(-X))
        
    w0=w0-alpha*slope0
    w1=w1-alpha*slope1

print(w0)
print(w1)
predicted_Y=w0+w1*X
plt.scatter(X[:,:1],y[:,:1])
plt.xlabel("Population")
plt.ylabel("Profit")
plt.scatter(X[:,:1],predicted_Y[:,:1])
```
