```python
import pandas as pn
import numpy as np
import math 
import random
```

```python
data=pn.read_csv("linear.csv",header=None)
data
```

```python
dataArray = np.array(data)
dataArray
```

```python
x=dataArray[:,:-1]
y=dataArray[:,1:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def linear_regression(x,y):
    w0=0.5
    w1=0.5
    length=len(x)/2
    alpha=.001
    j=[]

    for i in range(10000):
        j.append(sum((y-(w0+w1*x)**2)))
        predict_y =(w0+(w1*x))/length
        slope_w0 =sum((y-predict_y)*(-1))
        slope_w1 =sum((y-predict_y)*(-x))
        w0=w0-(alpha*slope_w0)
        w1=w1-(alpha*slope_w1)     
        
    return predict_y,j  
   
        
```

```python
predict_y,j=linear_regression(x,y)
```

```python
color='blue'
plt.scatter(x,y,c=color)
plt.scatter(x,predict_y,c='red',marker='.')
plt.xlabel("Population")
plt.ylabel("Profit")
   
```

```python
plt.plot([i for i in range(10000)],j)
plt.xlabel("iteration")
plt.ylabel("cost")
```
